## 数据可视化

```python
import numpy as np
import matplotlib.pyplot as plt

# 让图表直接在notebook中展示
%matplotlib inline
# 解决中文乱码问题 
plt.rcParams["font.sans-serif"] = 'simHei' 
# 解决负号无法正常显示的问题 
plt.rcParams['axes.unicode_minus'] = False 
# 将图表设置为矢量图格式，增加清晰度 
%config InlineBackend.figure_format = 'svg'
```

```python
# 绘制折线图
# 建立坐标系
plt.subplot(1, 1, 1)
# 指明x和y的值
x = np.arange(1, 10, 1)
y = np.array([866, 2335, 5710, 6482, 6120, 1605, 3813, 4428, 4631])
# 绘图
plt.plot(x, y, color='k', linestyle='dashdot', linewidth=1, marker='o', markersize=5, label='注册用户数')
# 设置标题
plt.title('1-9月注册用户量', loc='center')
# 添加数据标签
for a, b in zip(x, y):
    plt.text(a + 0.2, b, b, ha='left', va='bottom', fontsize=10)
# 设置网格线
plt.grid(True)
plt.legend()
# 保存到本地
# plt.savefig('file_path')
plt.show()
```

```python
# 绘制柱形图
plt.subplot(1, 1, 1)
x = np.array(["东区", "西区", "南区", "北区"])
y = np.array([8566, 6482, 5335, 7310])

plt.bar(x, y, width=0.5, align='center', label='任务量')
plt.title('全国各分区任务量', loc='center')

for a, b in zip(x, y):
    plt.text(a, b, b, ha='center', fontsize=12)

plt.xlabel('分区')
plt.ylabel('任务量')
plt.legend()
plt.show()
```

```python
# 绘制簇状柱形图
# 在同一张图绘制不同的y，错开x即可
plt.subplot(1, 1, 1)
x = np.array([1, 2, 3, 4])
y1 = np.array([8566, 5335, 7310, 6482])
y2 = np.array([4283, 2667, 3655, 3241])

plt.bar(x, y1, width=0.3, label='任务量')
plt.bar(x + 0.3, y2, width=0.3, label='完成量') # 相当于将完成量柱子左移0.3，正好是任务量柱子的宽度
plt.title("全国各区任务量和完成量", loc='center')

for a, b in zip(x, y1):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=12)
for a, b in zip(x + 0.3, y2):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=12)

plt.xlabel('区域')
plt.ylabel('任务情况')

plt.xticks(x + 0.15, ["东区", "南区", "西区", "北区"])
plt.legend()
plt.show()
```

```python
# 绘制堆积柱形图
# 在相同的x绘制不同的y即可自动叠加
plt.subplot(1, 1, 1)
x = ["东区", "南区", "西区", "北区"]
y1 = np.array([8566, 5335, 7310, 6482])
y2 = np.array([4283, 2667, 3655, 3241])

plt.bar(x, y1, width=0.5, label='任务量')
plt.bar(x, y2, width=0.5, label='完成量')
plt.title('全国各分区任务量和完成量', loc='center')

for a, b in zip(x, y1):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=12)
for a, b in zip(x, y2):
    plt.text(a, b, b, ha='center', va='top', fontsize=12)

plt.xlabel('区域')
plt.ylabel('任务情况')
plt.legend(loc='best', ncol=2)
plt.show()
```

```python
# 绘制条形图
plt.subplot(1, 1, 1)
x = ["东区", "南区", "西区", "北区"]
y = np.array([8566, 5335, 7310, 6482])

plt.barh(x, height=0.5, width=y, align='center')
plt.title('全国各分区任务量', loc='center')

for a, b in zip(x, y):
    plt.text(b, a, b, ha='left', va='center', fontsize=12)

plt.xlabel('区域')
plt.ylabel('任务情况')
plt.show()
```

```python
# 绘制散点图
plt.subplot(1, 1, 1)
x = np.random.randn(20)
y = np.random.randn(20)

plt.scatter(x, y, marker='o', s=100)
plt.title('随机散点图')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 绘制气泡图
plt.subplot(1, 1, 1)
x = np.random.randn(20)
y = np.random.randn(20)

plt.scatter(x, y, marker='o', s=abs(y) * 100)
plt.title('随机散点图')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 绘制面积图（堆积图）
plt.subplot(1, 1, 1)
x = np.arange(1, 10, 1)
y1 = np.array([866, 2335, 5710, 6482, 6120, 1605, 3813, 4428, 4631])
y2 = np.array([433, 1167, 2855, 3241, 3060, 802, 1906, 2214, 2315])

labels = ['注册人数', '激活人数']
plt.stackplot(x, y1, y2, labels=labels)
plt.title("1-9月注册与激活人数", loc='center')
plt.xlabel('月份')
plt.ylabel('注册与激活人数')
plt.legend()
plt.show()
```

```python
# 绘制饼图
plt.subplot(1, 1, 1)
x = np.array([8566, 5335, 7310, 6482])
labels = ['东区', '北区', '南区', '西区']
explode = [0.2, 0.05, 0.05, 0.05]
plt.pie(x, labels=labels, autopct='%.0f%%', shadow=True, explode=explode, radius=1, labeldistance=1.1)
plt.title('全国各区域任务量占比', loc='center')
plt.show()
```

```python
# 折线图 + 柱形图
plt.style.use('ggplot')
plt.subplot(1, 1, 1)
x = np.arange(1, 10, 1)
y1 = np.array([866, 2335, 5710, 6482, 6120, 1605, 3813, 4428, 4631])
y2 = np.array([433, 1167, 2855, 3241, 3060, 802, 1906, 2214, 2315])

plt.plot(x, y1, color='r', linestyle='solid', linewidth=1, marker='o', markersize=3, label='注册人数')
plt.bar(x, y2, color='b', label='激活人数')
plt.title('1-9月注册与激活人数', loc='center')

for a, b, c in zip(x, y1, y2):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=12)
    plt.text(a, c, c, ha='center', va='bottom', fontsize=12)

plt.xlabel('月份')
plt.ylabel('注册量')
plt.xticks(np.arange(1, 10, 1), ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月'])
plt.yticks(np.arange(1000, 7000, 1000), ['1000人', '2000人', '3000人', '4000人', '5000人', '6000人'])
plt.legend()
plt.show()
```

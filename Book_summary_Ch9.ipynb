```python
# create empty plot and subplots
fig = plt.figure() # plots are reset after each cell is evaluated
ax1 = fig.add_subplot(2, 2, 1) # create 2*2=4 subplots and select the first one
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(np.random.randn(50).cumsum(), 'k--') # plot on the last subplot by default
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3) # k means color black
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
# adjusting spacing
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, 
                    wspace=0, hspace=0) # adjust width and height between plots
```

```python
x = np.random.randn(30).cumsum()
plt.plot(x, 'g--')
plt.plot(x, linestyle='--', color='g') # equivalent to previous
plt.plot(np.random.randn(30).cumsum(), color='k', linestyle='dashed', marker='o')
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

```python
# Ticks
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
#ax.set_xlim([_start, _end])
#ax.set_ylim([_start, _end])
```

```python
# Legends
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best')
```

```python
# Seaborn
import seaborn as sns
tips = pd.read_csv('examples\Tips.txt')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
sns.barplot(x='tip_pct', y='day', data=tips, orient='h') # black line draws 95% confidence interval
```

```python
# Histograms
tips['tip_pct'].plot.hist(bins=50)
```

```python
# Density plot
tips['tip_pct'].plot.density()
```

```python
# Seaborn distplot
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))
sns.distplot(values, bins=100, color='k')
```

```python
# Scatter plot
macro = pd.read_csv('examples\macrodata.txt')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()

sns.regplot('m1', 'unemp', data=trans_data)
```

```python
# Pair plot
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

```python
# Facet plots for categorical data
sns.factorplot(x='day', y='tip_pct',  
               row='time', col='smoker', kind='bar', data=tips[tips.tip_pct<1]) # col means column
```

```python
sns.factorplot(x='day', y='tip_pct',  
               hue='time', col='smoker', kind='bar', data=tips[tips.tip_pct<1]) # within facet group
```

```python
sns.factorplot(x='tip_pct', y='day',  kind='box', data=tips[tips.tip_pct < 0.5]) 
```
